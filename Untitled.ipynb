```python
import json
import numpy as np
import pandas as pd
from functools import reduce

class NpDecoder(json.JSONDecoder):
    def __init__(self, *args, **kwargs):
        json.JSONDecoder.__init__(self, object_hook=self.dict_to_object, *args, **kwargs)

    def dict_to_object(self, d):
        if '__np_array__' in d:
            array_data = np.asarray(d['__np_array__'])
            return array_data
        return d
```

```python
# Leer el contenido del archivo "data_prior.json"
with open("data.json", "r") as file:
    content = file.read()

# Decodificar el contenido del archivo con NpDecoder
decoded_data = json.loads(content, cls=NpDecoder)
```

```python

list_dfs = []
for i,item in enumerate(decoded_data.items()):
    key = item[0]
    value = item[1]

    params = pd.DataFrame(value["params"])
    params["SERIES"] = [i]*params.shape[0]
    params["ENERGY"] = [value["energy"]]*params.shape[0]
    if "auth" in key:
        params["TYPE"] = ["AUTH"]*params.shape[0]
    if "non_auth" in key:
        params["TYPE"] = ["NON_AUTH"]*params.shape[0]
    list_dfs.append(params)
final_df = pd.concat(list_dfs).reset_index()
    
```

```python
def create_vectors(var):
    
    auth_list = []
    non_auth_list = []
    
    for energy in final_df["ENERGY"].unique():

        one_energy = final_df[final_df["ENERGY"] == energy]

        partial_ = one_energy[["index","ALPHA","BETA","PNB_VALUE","TYPE"]]
        for i in one_energy["index"].unique():
            partial_s_index = partial_[partial_["index"] == i]
            
            partial_s_index_auth = partial_s_index[var][partial_s_index["TYPE"]=="AUTH"]
            partial_s_index_no_auth = partial_s_index[var][partial_s_index["TYPE"]=="NON_AUTH"]
            


            auth_list.append(partial_s_index_auth.values[0])
            non_auth_list.append(partial_s_index_no_auth.values[0])
    return auth_list,non_auth_list

```

```python
alphas_destino,alphas_origen = create_vectors("ALPHA")
betas_destino,betas_origen = create_vectors("BETA")
pnb, _ = create_vectors("PNB_VALUE")
```

```python
u = np.array(alphas_destino) - np.array(alphas_origen)
v = np.array(betas_destino) - np.array(betas_origen)

import matplotlib.pyplot as plt 

fig = plt.figure(figsize=(10,8))
ax = fig.add_axes([0,0,1,1])
quiver = ax.quiver(alphas_origen, betas_origen, u, v,pnb, scale=1, scale_units='xy', angles='xy',alpha=.5)

ax.set_title("Parameter changes when the message is implemented",fontsize=20);
ax.set_xlim([0,1]);
ax.set_ylim([0,1]);
plt.xticks(fontsize=16, rotation=0)
plt.yticks(fontsize=16, rotation=0)
ax.grid()

ax.set_xlabel(r"$\alpha$",fontsize=25);
ax.set_ylabel(r"$\beta$",fontsize=25);

colorbar = plt.colorbar(quiver)
colorbar.ax.tick_params(labelsize=16)
colorbar.set_label('Personal normative value', fontsize=20)


plt.savefig("Evolution_w_message.jpg",bbox_inches="tight")
```

```python
slopes = [x/y for x,y in zip(u,v) if y != 0]

fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0,0,1,1])
ax.set_xlabel("Slope",fontsize=20)
plt.xticks(fontsize=16, rotation=0)
plt.yticks(fontsize=16, rotation=0)
plt.hist(slopes,bins=20,cumulative=True,density=True)
plt.grid()
plt.savefig("Slopes_histogram.jpg",bbox_inches="tight")
```
